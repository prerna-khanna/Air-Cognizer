```
src = cv2.imread('394.jpg') #Loading image 
src = cv2.resize(src,(960,540)) #Resizing the image 


dst = []  #Empty array for storing filtered image

d1 = 0   
x= 0
    
dst = cv2.Canny(src,50,100,3) #Detecting edeges in the source image 
 
for ix,iy in np.ndindex(dst.shape): 
    
    

    if(dst[ix,iy] > 0): #Passing those pixel values which are edges 

        x = x+1        #Incrementing X every time edge is detected 

        if( x == 270 ):  #Condition for reaching the threshold  

            d1 = ix      #Assigning the value of row to a varaible 
print(d1)
a = np.array(src)

if(d1 > 210):      #If the image has more than 40%  of sky 
    

    for iy in range (0,959):
        
        a[d1,iy] = 0     #Drawing a black line around horion 


    cv2.imshow('ans',a)   #Showing the results 

    cv2.waitkey(0)

else :
     print("Not appripriate for processing")
     for iy in range (0,959):
        
        a[d1,iy] = 0     #Drawing a black line around horion

        cv2.imshow('ans',a)

        cv2.waitkey(0)

        
```
